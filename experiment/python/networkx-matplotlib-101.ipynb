```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
def create_numeric_cities(num):
    for city in range(0, num):
        tsp.add_node(city)
```

```python
def create_random_edges(num):
    remaining = 0
    for _ in range(0, num):
        city1 = random.randrange(tsp.order())
        city2 = random.randrange(tsp.order())
        if city1 != city2 and not tsp.has_edge(city1, city2):
            tsp.add_edge(city1, city2)
        else:
            remaining += 1
    if remaining != num: # Avoid infinite loop 
        create_random_edges(remaining)
```

```python
def create_random_edges_each_node(connections):
    for city in tsp.nodes():
        for _ in range(connections):
            anotherCity = random.randrange(tsp.order())
            if city != anotherCity:
                tsp.add_edge(city, anotherCity)
```

```python
tsp = nx.Graph()
create_numeric_cities(15)
create_random_edges(15)
create_random_edges_each_node(2)
```

```python
print(nx.dijkstra_path(tsp, 1, tsp.order()-1), nx.dijkstra_path_length(tsp, 1, tsp.order()-1))
```

```python
nx.draw(tsp, pos=nx.circular_layout(tsp),with_labels=True)
```

```python
random_graph = nx.gnp_random_graph(15, 0.2)
nx.draw(random_graph, pos=nx.circular_layout(tsp),with_labels=True)
```

```python
connected = nx.complete_graph(5)
nx.draw(connected, pos=nx.circular_layout(tsp),with_labels=True)
```

```python
plt.title('- Graph -', fontsize=14, color='blue')
plt.ylabel(' Y-axis ', fontsize=14, color='blue')
plt.xlabel(' X-axis ', fontsize=14, color='blue')
plt.axis([0,60,1,20])
plt.plot(tsp.edges(), 'y-', tsp.nodes(), 'ro-', linewidth=0.5)
plt.show()
plt.scatter(tsp.nodes(), tsp.nodes())
plt.show()
plt.bar(tsp.nodes(), tsp.nodes())
```
